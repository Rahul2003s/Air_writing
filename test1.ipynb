```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
data=pd.read_csv('test.csv')
df=pd.DataFrame(data)
df.iloc[:,2]
plt.plot(np.arange(0,100),df.iloc[:,0][df.iloc[:,3] == 2][:100], label='X-axis')
plt.plot(np.arange (0,100),df.iloc[:,1][df.iloc[:,3] == 2][:100], label='Y-axis')
plt.plot(np.arange (0,100),df.iloc[:,2][df.iloc[:,3] == 2][:100], label='Z-axis')
plt.title("Graph")
plt.legend()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D



fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df.iloc[:,0],df.iloc[:,1],df.iloc[:,2])

# set axis labels and title
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')
ax.set_title('3D Scatter Plot')

# show the plot
plt.show()
```

```python

# generate some data for plotting
t = np.linspace(0, 10, 100)
x= df.iloc[:,0]
y = df.iloc[:,1]
z = df.iloc[:,2]

# create a 3D figure
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# plot the line
ax.plot(x, y, z)

# set axis labels
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# show the plot
plt.show()
```

```python
fig = plt.figure(figsize = (8,8))
ax = plt.axes(projection='3d')
ax.grid()
t = np.arange(0, 10*np.pi, np.pi/50)
x = np.sin(t)
y = np.cos(t)

ax.plot3D(x, y, t)
ax.set_title('3D Parametric Plot')

# Set axes label
ax.set_xlabel('x', labelpad=20)
ax.set_ylabel('y', labelpad=20)
ax.set_zlabel('t', labelpad=20)

plt.show()
```
